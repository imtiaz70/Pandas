## DataFrame Basic Functionality

```python
import numpy as np
import pandas as pd

dataOfEmp = pd.DataFrame({'Age':pd.Series([-23,42,41,45,54]) , 'Name':pd.Series(['Tom' , 'James' , 'Ricky' , 'Vin' , 'Steve']) , 'Rating':pd.Series([4.23,3.89,3.24,2.59,3.20]) } )
print(dataOfEmp)
# print("\n abs of DataFrame \n" , dataOfEmp.abs(0,2))
print("\n Prod of dataframe \n" , dataOfEmp.prod())

print("\n Describe of dataframe \n" , dataOfEmp.describe(include='all'))
print("\n minimum of DataFrame \n" , dataOfEmp.min(0))
print("\n maximum of DataFrame \n" , dataOfEmp.max(0))
print("\n Mode of DataFrame \n" , dataOfEmp.mode(0))
print("\n Median of DataFrame \n" , dataOfEmp.median(0))
print("\n count of DataFrame \n" , dataOfEmp.count())
print("\n Sum of DataFrame \n" , dataOfEmp.sum(0))
print("\n CumSum of DataFrame \n" , dataOfEmp.cumsum(0))
print("\n Standered Deviation of DataFrame \n" , dataOfEmp.std(0))
print("\n Mean of DataFrame \n" , dataOfEmp.mean(0))
print("\n Head of DataFrame \n" , dataOfEmp.head(2))
print("\n Tail of DataFrame \n" , dataOfEmp.tail(2))
print("\n Values of DataFrame \n" , dataOfEmp.values)
print("Transpose: \n" , dataOfEmp.T)
print("\n Axis Are \n" , dataOfEmp.axes)
print("\n Data type \n " , dataOfEmp.dtypes)
print("\n Check DataFrame is Empty \n " , dataOfEmp.empty)
print("\n  Dimesntion of DataFrame is \n" , dataOfEmp.ndim)
print("\n Shape of DataFrame \n" , dataOfEmp.shape)
print("\n Size of DataFrame \n" , dataOfEmp.size)

```
